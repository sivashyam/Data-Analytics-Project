```python
x=df[['Unit_Price','Unit_Cost','Units_Sold']]
x
```

```python
y=df[['Total_Profit']]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=1,random_state=50)
print(x_train,x_test,y_train,y_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaled=StandardScaler()
x_scaled=scaled.fit_transform(x)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
```

```python
#linear regression
model=LinearRegression()
model.fit(x_train,y_train)
model
```

```python
ypred=model.predict(x_test)
ypred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse=mean_squared_error(y_test,ypred)
print(mse)
```

```python
mae=mean_absolute_error(y_test,ypred)
print(mae)
```

```python
r2score=r2_score(y_test,ypred)
print(r2score)
```

```python
# New sales input: unit sold,unit price,unit cost
ne = np.array([[7542,152.58,97.44]])

# Predict
ps = model.predict(ne)

# Print properly
print(f"Predicted Total_profit: {ps[0].item():.2f}")
```
